```python
%matplotlib notebook
import control
import numpy as np
import scipy as sp
from scipy import signal
from matplotlib import pylab as plt
from ipywidgets import *
from IPython.display import display
from IPython.html import widgets
plt.style.use('ggplot')
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
def polymul(arrays):
    salida = [1]
    for i in arrays:
        salida = np.polymul(salida,i)
    return salida
```

## Ej1 a

```python
sysA1 = signal.lti([1,0.1],polymul(([1,0],[1,2],[1,0.01])));
control.root_locus(sysA1, kvect= np.linspace(0.001,100,10000));
plt.title("Ejercicio A1")
```

```python
sysA2 = signal.lti([1,0.1],polymul(([1,0.01],[1,0],[1,2],[1,5])));
control.root_locus(sysA2, kvect= np.logspace(-3,2,10000));
plt.title("Ejercicio A2")
```
