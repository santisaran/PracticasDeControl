## Ej1 a

```python
def Ej1(num,den):
    sys1 = signal.lti(num,den) # Creamos el sistema
    ctf = control.tf(num,den) 
    w, mag, phase = signal.bode(sys1) # Diagrama de bode: frecuencias, magnitud y fase
    figb, (axb1, axb2) = plt.subplots(2, 1)
    axb1.semilogx(w, mag) # Eje x logarítmico
    axb2.semilogx(w, phase) # Eje x logarítmico
    axb1.grid(True,'major')
    axb2.grid(True,'minor')
    w, H = signal.freqresp(sys1)
    fign, (axn1, axn2) = plt.subplots(1, 2)
    axn1.plot(H.real, H.imag)
    axn1.plot(H.real, -H.imag)
    axn2.plot(sys1.zeros.real, sys1.zeros.imag, 'o')
    axn2.plot(sys1.poles.real, sys1.poles.imag, 'x')
    figni, (axni1) = plt.subplots(1, 1)
    control.nichols_plot(ctf)
def Ej1Control(num,den,omega=None):
    
    figb, (axb1, axb2) = plt.subplots(1, 2)
    control.bode_plot(control.tf(num,den),omega);
    fign, (axn1) = plt.subplots(1, 1)
    control.nyquist_plot(control.tf(num,den),omega);
    figni, (axni1) = plt.subplots(1, 1)
    control.nichols_plot(control.tf(num,den),omega);
```

```python
Ej1Control([40], [1 , 2.5, 1],omega = np.logspace(-3,3,500))
#ret = control.feedback(control.tf([40], [1 , 2.5, 1]))
#Ej1Control(ret.num,ret.den)
```
